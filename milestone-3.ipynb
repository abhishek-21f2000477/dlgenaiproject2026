```python
# ===============================
# Import libraries
# ===============================
import torch
import torch.nn as nn
import torchaudio
import math

print("---- Q1 Flatten Tensor ----")

# Q1: Flatten tensor excluding batch dimension
x = torch.randn(32, 1, 64, 64)
flattened_features = x.view(x.size(0), -1).shape[1]
print("Input features for Linear layer:", flattened_features)


print("\n---- Q2 Mel Spectrogram Shape ----")

# Q2: Mel Spectrogram shape
sample_rate = 16000
duration = 1  # seconds
audio = torch.randn(1, sample_rate * duration)

mel_transform = torchaudio.transforms.MelSpectrogram(
    sample_rate=16000,
    n_fft=400,
    hop_length=160,
    n_mels=64
)

mel_spec = mel_transform(audio)
print("Mel Spectrogram shape:", tuple(mel_spec.shape))


print("\n---- Q3 CNN Output Shape ----")

# Q3: Conv + MaxPool output size
conv = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=3, stride=1, padding=1)
pool = nn.MaxPool2d(kernel_size=2, stride=2)

x = torch.randn(1, 1, 64, 64)

x = conv(x)
x = pool(x)

print("Output spatial shape:", x.shape[2:])


print("\n---- Q4 Linear Layer Parameters ----")

# Q4: Number of parameters in Linear layer
linear = nn.Linear(128, 64)

params = sum(p.numel() for p in linear.parameters())
print("Total learnable parameters:", params)


print("\n---- Q5 Number of Batches ----")

# Q5: batches in one epoch
dataset_size = 1050
batch_size = 32

batches = math.ceil(dataset_size / batch_size)
print("Total batches per epoch:", batches)


print("\n---- Q6 CrossEntropy Loss ----")

# Q6: Cross entropy loss calculation
logits = torch.tensor([[2.5, 1.0, 0.1]])
target = torch.tensor([0])

criterion = nn.CrossEntropyLoss()
loss = criterion(logits, target)

print("CrossEntropyLoss:", round(loss.item(), 3))


print("\n---- Q7 SGD Weight Update ----")

# Q7: SGD update
weight = torch.tensor(0.5)
gradient = 0.2
lr = 0.01

new_weight = weight - lr * gradient

print("New weight:", round(new_weight.item(), 3))


print("\n---- Q8 Same Padding ----")

# Q8: padding for same output
kernel_size = 5
padding = (kernel_size - 1) // 2

print("Required padding:", padding)


print("\n---- Q9 WandB Logging Command ----")

# Q9: Example logging command
print('wandb.log({"loss": loss, "acc": accuracy})')


print("\n---- Q10 Accuracy ----")

# Q10: accuracy
correct = 170
total = 200

accuracy = correct / total
print("Accuracy:", accuracy)
```
